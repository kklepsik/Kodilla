Zadanie: statystyki
Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)

data_100 = pd.DataFrame(data_100, columns=['X'])
data_10000 = pd.DataFrame(data_10000, columns=['X'])
data_100.head()
```

```python
print(data_100.describe())
print(data_10000.describe())
```

```python
figure, axes = plt.subplots(1, 2)
axes[0].hist(data_100['X'], bins=20)
axes[1].hist(data_10000['X'], bins=20)
plt.show()
```

```python
plt.figure(figsize = (8, 6))

plt.subplot(1, 2, 1)                # wiersz, kolumna, numer aktywowanego wykresu
sns.boxplot(data_100['X'])

plt.subplot(1, 2, 2)
sns.boxplot(data_10000['X'])
```
