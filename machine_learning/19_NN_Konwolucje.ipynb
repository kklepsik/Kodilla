```python
# ===============================
# #0. Importy – biblioteki i funkcje
# ===============================
import os
import zipfile
import random
import shutil

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, UnidentifiedImageError

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, BatchNormalization, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2

# ===============================
# #1. Przygotowanie danych – pobranie, filtracja, podział
# ===============================
# Rozpakuj archiwum z dysku lokalnego
with zipfile.ZipFile(r'C:\Python\Kodilla\MACHINE_LEARNING\kagglecatsanddogs_5340.zip','r') as z:
    z.extractall('C:/tmp')

# Funkcja wybierająca tylko poprawne pliki JPG
def get_valid(path):
    valid=[]
    for f in os.listdir(path):
        try:
            Image.open(os.path.join(path,f))
            valid.append(f)
        except UnidentifiedImageError:
            pass
    return valid

# Ścieżki oryginalne
cat_src = 'C:/tmp/PetImages/Cat'
dog_src = 'C:/tmp/PetImages/Dog'
cats = get_valid(cat_src)
dogs = get_valid(dog_src)
random.seed(101)
random.shuffle(cats)
random.shuffle(dogs)

# Wyznacz wielkości zbiorów
n = min(len(cats), len(dogs))
train_n = int(0.7 * n)
valid_n = int(0.2 * n)
test_n  = n - train_n - valid_n

# Stwórz foldery dla train/valid/test pod klasami
base = 'dataset'
for split in ['train','valid','test']:
    for cls in ['Cat','Dog']:
        os.makedirs(f'{base}/{split}/{cls}', exist_ok=True)

# Rozdziel i kopiuj pliki do odpowiednich folderów
for i, (c, d) in enumerate(zip(cats[:n], dogs[:n])):
    if i < train_n:
        split = 'train'
    elif i < train_n+valid_n:
        split = 'valid'
    else:
        split = 'test'
    shutil.copyfile(f'{cat_src}/{c}', f'{base}/{split}/Cat/{c}')
    shutil.copyfile(f'{dog_src}/{d}', f'{base}/{split}/Dog/{d}')

# ===============================
# #2. Generatory – rescale + augmentacja
# ===============================
img_h = img_w = 150
batch_size = 64

# Tylko normalizacja pikseli [0,1]
datagen = ImageDataGenerator(rescale=1./255)

# Normalizacja + augmentacja (dla modeli, które tego potrzebują)
aug = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=30,
    horizontal_flip=True
)

# Generowanie batchy – poprawne użycie nazwanych argum entów
train_gen = datagen.flow_from_directory(
    directory=f'{base}/train',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='binary'
)
valid_gen = datagen.flow_from_directory(
    directory=f'{base}/valid',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='binary'
)
train_aug_gen = aug.flow_from_directory(
    directory=f'{base}/train',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='binary'
)

steps_per_epoch  = (train_n * 2) // batch_size
validation_steps = (valid_n * 2) // batch_size

# ===============================
# #3. Funkcja do rysowania historii treningu
# ===============================
def plot_history(h, name):
    df = pd.DataFrame(h.history)
    epochs = df.index + 1
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(epochs, df['accuracy'], 'b-', label='train')
    plt.plot(epochs, df['val_accuracy'], 'r-', label='val')
    plt.title(f'{name} – Accuracy'); plt.legend()
    plt.subplot(1,2,2)
    plt.plot(epochs, df['loss'], 'b-', label='train')
    plt.plot(epochs, df['val_loss'], 'r-', label='val')
    plt.title(f'{name} – Loss'); plt.legend()
    plt.tight_layout()
    plt.show()

# ===============================
# #4. Definicje modeli prostych
# ===============================
def build_baseline():
    # #baseline: bez warstw splotowych
    m = Sequential([
        Flatten(input_shape=(img_h,img_w,3)),
        Dense(1, activation='sigmoid')
    ])
    m.compile(loss='binary_crossentropy', optimizer=RMSprop(1e-4), metrics=['accuracy'])
    return m

def build_simple(pad):
    # #simple: jedna warstwa Conv + pooling
    m = Sequential([
        Conv2D(10,3,activation='relu',padding=pad, input_shape=(img_h,img_w,3)),
        MaxPooling2D(),
        Flatten(),
        Dense(1, activation='sigmoid')
    ])
    m.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return m

# ===============================
# #5. Nowe modele – dodatkowe warstwy splotowe + regularyzacja
# ===============================
def build_extra_1():
    # #extra_1: 4 bloki Conv→BatchNorm→Pool + Dropout+L2 w Dense
    m = Sequential()
    for f in [16,32,64,128]:
        if not m.layers:
            m.add(Conv2D(f,3,padding='same',activation='relu',input_shape=(img_h,img_w,3)))
        else:
            m.add(Conv2D(f,3,padding='same',activation='relu'))
        m.add(BatchNormalization())
        m.add(MaxPooling2D())
    m.add(Flatten())
    m.add(Dense(256, activation='relu', kernel_regularizer=l2(0.005)))
    m.add(Dropout(0.5))
    m.add(Dense(128, activation='relu', kernel_regularizer=l2(0.005)))
    m.add(Dropout(0.5))
    m.add(Dense(1, activation='sigmoid'))
    m.compile(loss='binary_crossentropy', optimizer=RMSprop(1e-4), metrics=['accuracy'])
    return m

def build_extra_2():
    # #extra_2: jak extra_1 + L2 regularyzacja w Conv2D
    m = Sequential()
    for f in [16,32,64,128]:
        if not m.layers:
            m.add(Conv2D(f,3,padding='same',activation='relu',
                         kernel_regularizer=l2(0.001),
                         input_shape=(img_h,img_w,3)))
        else:
            m.add(Conv2D(f,3,padding='same',activation='relu',kernel_regularizer=l2(0.001)))
        m.add(BatchNormalization())
        m.add(MaxPooling2D())
    m.add(Flatten())
    m.add(Dense(256, activation='relu', kernel_regularizer=l2(0.005)))
    m.add(Dropout(0.5))
    m.add(Dense(128, activation='relu', kernel_regularizer=l2(0.005)))
    m.add(Dropout(0.5))
    m.add(Dense(1, activation='sigmoid'))
    m.compile(loss='binary_crossentropy', optimizer=RMSprop(1e-4), metrics=['accuracy'])
    return m

# ===============================
# #6. Trenowanie – wszystkie modele
# ===============================
models = {
    'baseline': build_baseline(),
    'simple_1': build_simple('valid'),
    'simple_2': build_simple('same'),
    'extra_1': build_extra_1(),   # NOWY: dodatkowe warstwy i regularyzacja
    'extra_2': build_extra_2(),   # NOWY: L2 w Conv2D + dodatkowy blok
}

histories = {}
for name, model in models.items():
    print(f'\n### Trenowanie modelu: {name}')
    es = EarlyStopping(patience=5, monitor='val_accuracy', restore_best_weights=True)
    ck = ModelCheckpoint(f'{name}.keras', save_best_only=True, monitor='val_accuracy')
    gen = train_aug_gen if name in ('extra_1','extra_2') else train_gen
    history = model.fit(
        gen,
        steps_per_epoch=steps_per_epoch,
        validation_data=valid_gen,
        validation_steps=validation_steps,
        epochs=100,
        callbacks=[es, ck],
        verbose=1
    )
    histories[name] = history
    plot_history(history, name)

# ===============================
# #7. Ewaluacja nowych modeli na teście
# ===============================
test_gen = datagen.flow_from_directory(
    directory=f'{base}/test',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)
for name in ('extra_1','extra_2'):
    m = load_model(f'{name}.keras')
    loss, acc = m.evaluate(test_gen, verbose=0)
    print(f'{name}: test_accuracy = {acc:.4f}')
```
