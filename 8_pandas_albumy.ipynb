Wczytanie bazy danych

```python
import pandas as pd
import lxml         #tego nie było w kursie, ale był błąd

albumy = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551', header = 0)
dataframe = albumy[0]
```

Edycja tabeli

```python
dataframe.rename(columns={'POS': 'Pozycja', 'TITLE': 'Tytuł', 'ARTIST': 'Artysta', 'YEAR': 'Rok', 'HIGH POSN': 'Max POZ'}, inplace=True)
dataframe
```

```python
dataframe['Artysta'].value_counts()             #najczęściej pojawiali się TAKE THAT i COLDPLAY
```

```python
ile_unikatow = len(dataframe['Artysta'].unique())           #liczba artystów
ile_unikatow
```

```python
dataframe.columns = dataframe.columns.str.capitalize()          #wielkie litery
dataframe
```

```python
dataframe.drop('Max poz', axis = 1, inplace = True)             #usuniecie kolumny
dataframe

#po usunięciu błąd, bo kolumna została zastąpiona
```

```python
top_rok =dataframe['Rok'].value_counts(normalize = True)        #Najbardziej popularny rok

top_1 = top_rok.iloc[0]
top_rok[top_rok == top_1]
```

```python
dataframe[(dataframe['Rok'] >= 1960) & (dataframe['Rok'] <= 1990)]        #Albumy wyprodukowane w latach 1960-1990  
#dataframe['Rok'].dtype
```

```python
dataframe['Rok'].max()              #najwcześniejszy rok albumu

rok_posortowane = dataframe.sort_values(by = 'Rok', ascending = False)      #czy jest prostsza metoda?
rok_posortowane.iloc[0]


```

```python
lista = rok_posortowane.drop_duplicates(subset = 'Artysta', keep = 'last')      #lista najwcześniej wydanych albumów

#dataframe.sort_values(by = 'Artysta')
moja_lista = 'moja_lista.csv'
lista.to_csv(moja_lista, index = False)

import os

path = os.path.abspath(moja_lista)
print(f"Plik został zapisany w: {path}")

df = pd.read_csv(path)
df = df[['Artysta', 'Tytuł', 'Rok', 'Pozycja']]
df2 = df.drop(['Pozycja'], axis = 1)
df2
```
